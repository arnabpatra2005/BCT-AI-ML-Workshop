if-else statement

```python
# if-else statement & identation

# Let's check you're eligible for driving or not

age=int(input("Enter Your Age: "))
if age>18:
  print("You're eligible to drive")
else:
  print("You're not eligibke to drive")
```

```python
# if-else statement & identation

# Let's find the greatest among three numbers
# logical operators: and, or, not

a=int(input("Enter a: "))
b=int(input("Enter b: "))
c=int(input("Enter c: "))

if a>b and b>c:
  print("a is the greatest number")
elif b>a and b>c:
  print("b is the greatest number")
else:
  print("c is the greatest number")
```
