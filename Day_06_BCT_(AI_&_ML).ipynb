# # **Project 9 : Employee Dataset Creation**

```python
import pandas as pd

data = {
    'Employee Name': [
        'Arnab', 'Arpan', 'Ankush', 'Sanju',
    ],
    'Employee ID': [1001, 1002, 1003, 1004],
    'Employee Salary': [85000, 72000, 93000, 110000],
    'Employee Feedback': ['Good','Very Good','Excelent','Good']
}

df = pd.DataFrame(data)

print("Dataset Created Successfully!")
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

one = OneHotEncoder(sparse_output=False)
le = LabelEncoder()

df["Employee Feedback"] = le.fit_transform(df["Employee Feedback"])

name_encoded = one.fit_transform(df[['Employee Name']])

name_df = pd.DataFrame(name_encoded, columns=one.get_feature_names_out(['Employee Name']))

df_final = pd.concat([df.drop('Employee Name', axis=1), name_df], axis=1)

df_final.head()
```
